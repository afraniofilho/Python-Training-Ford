```python
import yfinance as yf
```

```python
import numpy as np
import pandas as pd
```

```python
ticker = "USDBRL=X"
```

```python
carteira = yf.download(ticker, start='2021-01-01')
```

```python
carteira.head()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
```

```python
type(fig)
```

```python
!pip install --upgrade plotly
```

```python
fig.add_trace(go.Scatter(x=list(carteira.index), y=list(carteira.Close)))
fig.add_trace(go.Scatter(x=list(carteira.index), y=list(carteira.Low)))
fig.update_layout(title_text="Cotação do Dólar vs Real")
```

```python
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
fig.show()
```
