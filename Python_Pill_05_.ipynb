```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
df = pd.read_csv('2102.csv', sep=';', skiprows=2)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.loc[df['Público'] != "ND"]
```

```python
df = df.loc[df['Renda (R$)'] != "ND"]
```

```python
df['Público'] = pd.to_numeric(df['Público'])
```

```python
df['Renda (R$)'] = pd.to_numeric(df['Renda (R$)'])
```

```python
# Quantidade de Público total por ano de lançamentos
```

```python
df_publico = df.groupby('Ano de Lançamento')['Público'].sum()
```

```python
df_publico
```

```python
fig = px.bar(df_publico, x=df_publico.index, y='Público', title='Público por Ano ')
fig.show()
```

```python
fig = px.bar(df, x='Ano de Lançamento', y='Público', title='Público por Ano', color='Renda (R$)')
fig.show()
```

```python
fig = px.bar(df, x='Ano de Lançamento', y='Público', title='Público por Ano', color='Renda (R$)', text='Título')
fig.show()
```
